## Time Series EDA

```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pdr.get_data_yahoo("MSFT")
```

```python
data.head()
```

```python
data["High"].plot(figsize=(12, 4))
```

```python
data["Low"].plot(figsize=(12,4))
```

```python
data["Close"].plot(figsize=(12, 4))
```

```python
data["Open"].plot(xlim=["2021-12-01", "2022-09-14"], ylim=[230,350],figsize=(12, 4))
```

```python
data["Low"].plot(xlim=["2021-12-01", "2022-09-14"], ylim=[230,350],figsize=(12, 4))
```

```python
index = data.loc["2021-09-01":"2022-09-14"].index
```

```python
index
```

```python
share_open = data.loc["2021-09-01":"2022-09-14"]["Open"]
```

```python
share_open
```

```python
figure, axis = plt.subplots(figsize=(12,4))
axis.plot(index, share_open)
```

```python
data.resample(rule="A").max()
```

```python
high_open_price_year = data["Open"].resample(rule="A").max()
```

```python
high_open_price_year
```

```python
data["Open"].resample(rule="A").max().plot(kind="bar")
```

```python
low_close_price_year = data["Close"].resample(rule="A").min()
```

```python
low_close_price_year
```

```python
data["Close"].resample(rule="A").min().plot(kind="bar")
```

```python
month_data = data.resample(rule="M").max()
```
